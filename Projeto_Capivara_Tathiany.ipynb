Exercício 5: Novos Clientes e Expansão do Negócio

```python
"""
Calcula o total de novos clientes para o DataFrame da região Norte.
"""
total_novos_norte = df_norte['Novos Clientes'].sum()

"""
Calcula o total de novos clientes para o DataFrame da região Sul.
"""
total_novos_sul = df_sul['Novos Clientes'].sum()

"""
Calcula o total de novos clientes para o DataFrame da região Leste.
"""
total_novos_leste = df_leste['Novos Clientes'].sum()

"""
Calcula o total de novos clientes para o DataFrame da região Oeste.
"""
total_novos_oeste = df_oeste['Novos Clientes'].sum()
```

```python
"""
Cria um dicionário que mapeia cada região ao seu total de novos clientes.
"""
totais_novos_clientes = {
    'Norte': total_novos_norte,
    'Sul': total_novos_sul,
    'Leste': total_novos_leste,
    'Oeste': total_novos_oeste
}

"""
Identifica a região com o maior número de novos clientes.
A função `max` é usada com o parâmetro `key` para encontrar a região com o valor máximo no dicionário.
"""
regiao_mais_clientes_novos = max(totais_novos_clientes, key=totais_novos_clientes.get)

"""
Exibe a região com o maior número de novos clientes.
"""
print("Região com mais novos clientes:", regiao_mais_clientes_novos)
```

```python
"""
Agrupa o DataFrame pela coluna 'Região' e calcula a soma de novos clientes para cada região.
O método `reset_index()` transforma o resultado em um DataFrame.
Em seguida, ordena o DataFrame pela coluna 'Novos Clientes' em ordem decrescente.
"""
clientes_por_regiao = df.groupby("Região")["Novos Clientes"].sum().reset_index()
clientes_por_regiao = clientes_por_regiao.sort_values(by="Novos Clientes", ascending=False)

"""
Agrupa o DataFrame pela coluna 'Região' e calcula a soma das colunas 'Receita' e 'Lucro' para cada região.
O método `reset_index()` transforma o resultado em um DataFrame.
Em seguida, ordena o DataFrame pela coluna 'Receita' em ordem decrescente.
"""
faturamento_lucro_por_regiao = df.groupby("Região")[["Receita", "Lucro"]].sum().reset_index()
faturamento_lucro_por_regiao = faturamento_lucro_por_regiao.sort_values(by="Receita", ascending=False)

"""
Cria um gráfico de barras para visualizar o faturamento e o lucro por região.
- As barras azuis representam o faturamento.
- As barras vermelhas representam o lucro.
- O parâmetro `alpha` define a transparência das barras.
- O eixo X representa as regiões.
- O eixo Y representa os valores de faturamento e lucro.
- O gráfico é rotacionado em 45 graus para melhor visualização.
- A legenda é adicionada para diferenciar faturamento e lucro.
"""
plt.figure(figsize=(10,6))
plt.bar(faturamento_lucro_por_regiao["Região"], faturamento_lucro_por_regiao["Receita"], color='pink', alpha=0.6, label='Faturamento')
plt.bar(faturamento_lucro_por_regiao["Região"], faturamento_lucro_por_regiao["Lucro"], color='purple', alpha=0.6, label='Lucro')
plt.xlabel("Região")
plt.ylabel("Valores")
plt.title("Faturamento e Lucro por Região")
plt.xticks(rotation=45)
plt.legend()
plt.show()

"""
Exibe o DataFrame `faturamento_lucro_por_regiao` sem o índice.
"""
print(faturamento_lucro_por_regiao.to_string(index=False))
```

```python
"""
Converte a coluna 'Data' para o tipo datetime, permitindo manipulações de data e hora.
"""
df['Data'] = pd.to_datetime(df['Data'])

"""
Agrupa o DataFrame pela coluna 'Data' e calcula a soma da coluna 'Quantidade de Vendas' para cada dia.
O resultado é uma série onde o índice são as datas e os valores são as quantidades de vendas.
"""
vendas_por_dia = df.groupby('Data')['Quantidade de Vendas'].sum()

"""
Cria um gráfico de linha para visualizar a quantidade de vendas por dia.
- O eixo X representa as datas.
- O eixo Y representa a quantidade de vendas.
- O gráfico é criado usando a biblioteca Seaborn para melhor visualização.
- O título do gráfico é "Quantidade de Vendas por Dia".
"""
plt.figure(figsize=(25, 6))
sns.lineplot(x=vendas_por_dia.index, y=vendas_por_dia.values)
plt.title('Quantidade de Vendas por Dia')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.show()

"""
Calcula a média e o desvio padrão das vendas por dia.
"""
media_vendas = vendas_por_dia.mean()
desvio_padrao_vendas = vendas_por_dia.std()

"""
Define limites para identificar dias com vendas significativamente altas ou baixas:
- Limite superior: média + 2 * desvio padrão.
- Limite inferior: média - 1 * desvio padrão.
"""
limite_superior = media_vendas + 2 * desvio_padrao_vendas
limite_inferior = media_vendas - desvio_padrao_vendas

"""
Identifica os dias com vendas significativamente altas (acima do limite superior).
"""
dias_mais_vendas = vendas_por_dia[vendas_por_dia > limite_superior]

"""
Identifica os dias com vendas significativamente baixas (abaixo do limite inferior).
"""
dias_menos_vendas = vendas_por_dia[vendas_por_dia < limite_inferior]

"""
Exibe as datas com aumento significativo de vendas.
"""
print('Datas com aumento significativo de vendas:')
print(dias_mais_vendas)

"""
Exibe as datas com diminuição significativa de vendas.
"""
print('Datas com diminuição significativa de vendas:')
print(dias_menos_vendas)
```

```python
"""
Converte a coluna 'Data' para o tipo datetime, permitindo manipulações de data e hora.
"""
df['Data'] = pd.to_datetime(df['Data'])

"""
Agrupa o DataFrame pela coluna 'Data' e calcula a soma da coluna 'Quantidade de Vendas' para cada dia.
O resultado é uma série onde o índice são as datas e os valores são as quantidades de vendas.
"""
vendas_por_dia = df.groupby('Data')['Quantidade de Vendas'].sum()

"""
Cria um gráfico de linha para visualizar a quantidade de vendas por dia.
- O eixo X representa as datas.
- O eixo Y representa a quantidade de vendas.
- O gráfico é criado usando a biblioteca Seaborn para melhor visualização.
- O título do gráfico é "Quantidade de Vendas por Dia".
"""
plt.figure(figsize=(25, 6))
sns.lineplot(x=vendas_por_dia.index, y=vendas_por_dia.values)
plt.title('Quantidade de Vendas por Dia')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.show()

"""
Calcula a média e o desvio padrão das vendas por dia.
"""
media_vendas = vendas_por_dia.mean()
desvio_padrao_vendas = vendas_por_dia.std()

"""
Define limites para identificar dias com vendas significativamente altas ou baixas:
- Limite superior: média + 2 * desvio padrão.
- Limite inferior: média - 1 * desvio padrão.
"""
limite_superior = media_vendas + 2 * desvio_padrao_vendas
limite_inferior = media_vendas - desvio_padrao_vendas

"""
Identifica os dias com vendas significativamente altas (acima do limite superior).
"""
dias_mais_vendas = vendas_por_dia[vendas_por_dia > limite_superior]

"""
Identifica os dias com vendas significativamente baixas (abaixo do limite inferior).
"""
dias_menos_vendas = vendas_por_dia[vendas_por_dia < limite_inferior]

"""
Exibe as datas com aumento significativo de vendas.
"""
print('Datas com aumento significativo de vendas:')
print(dias_mais_vendas)

"""
Exibe as datas com diminuição significativa de vendas.
"""
print('Datas com diminuição significativa de vendas:')
print(dias_menos_vendas)
```

Pergunta: Alguma região tem um desempenho excepcional para certos produtos?


```python
"""
Agrupa o DataFrame pelas colunas 'Região' e 'Categoria do Produto' e calcula a soma da coluna 'Quantidade de Vendas' para cada grupo.
O método `reset_index()` transforma o resultado em um DataFrame.
"""
produtos_por_regiao = df.groupby(["Região", "Categoria do Produto"]).agg({"Quantidade de Vendas": "sum"}).reset_index()

"""
Transforma o DataFrame em uma tabela de heatmap, onde:
- As linhas são as categorias de produtos.
- As colunas são as regiões.
- Os valores são as quantidades de vendas.
Os valores ausentes são preenchidos com 0.
"""
tabela_heatmap = produtos_por_regiao.pivot(index="Categoria do Produto", columns="Região", values="Quantidade de Vendas").fillna(0)

"""
Cria um heatmap para visualizar o desempenho dos produtos por região.
- O mapa de cores 'Blues' é usado para representar os valores.
- Os valores são anotados no heatmap com formatação de números inteiros (`fmt=".0f"`).
- As linhas entre as células têm largura de 0.5.
- O eixo X representa as regiões.
- O eixo Y representa as categorias de produtos.
- O título do gráfico é "Desempenho de Produtos por Região".
- Os rótulos do eixo X são rotacionados em 45 graus para melhor visualização.
"""
plt.figure(figsize=(12,6))
sns.heatmap(tabela_heatmap, cmap="pink", annot=True, fmt=".0f", linewidths=0.5)
plt.xlabel("Região")
plt.ylabel("Produto")
plt.title("Desempenho de Produtos por Região")
plt.show()

"""
Identifica os produtos com melhor desempenho (maior quantidade de vendas) em cada região.
- O método `groupby` agrupa os dados por região.
- O método `idxmax` retorna o índice da linha com o maior valor de 'Quantidade de Vendas' para cada grupo.
- O método `loc` é usado para selecionar as linhas correspondentes.
"""
print("Produtos com melhor desempenho por região:")
melhores_produtos = produtos_por_regiao.loc[produtos_por_regiao.groupby("Região")["Quantidade de Vendas"].idxmax()]
print(melhores_produtos.to_string(index=False))
```

```python
"""
Agrupa o DataFrame pela coluna 'Categoria do Produto' e calcula:
- A soma da coluna 'Quantidade de Vendas'.
- A média da coluna 'Satisfação do Cliente'.
O método `reset_index()` transforma o resultado em um DataFrame.
"""
produtos_vendas_avaliacao = df.groupby("Categoria do Produto").agg({"Quantidade de Vendas": "sum", "Satisfação do Cliente": "mean"}).reset_index()

"""
Cria um gráfico de barras e linha para visualizar a relação entre a quantidade de vendas e a satisfação média por produto.
- O eixo X representa as categorias de produtos.
- O eixo Y à esquerda representa a quantidade de vendas (barras azuis).
- O eixo Y à direita representa a satisfação média (linha vermelha).
- O gráfico é criado usando dois eixos Y (`ax1` e `ax2`).
"""
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()

ax1.bar(produtos_vendas_avaliacao["Categoria do Produto"], produtos_vendas_avaliacao["Quantidade de Vendas"], color='blue', alpha=0.6, label='Quantidade de Vendas')

ax2.plot(produtos_vendas_avaliacao["Categoria do Produto"], produtos_vendas_avaliacao["Satisfação do Cliente"], color='red', marker='o', linestyle='-', label='Satisfação Média')

ax1.set_xlabel("Produto")
ax1.set_ylabel("Quantidade de Vendas", color='blue')
ax2.set_ylabel("Média de Satisfação", color='red')

plt.title("Relação entre Quantidade de Vendas e Satisfação Média")
plt.xticks(rotation=45)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()

"""
Calcula a correlação de Pearson entre a quantidade de vendas e a satisfação média.
- `correlacao_satisfacao`: Valor da correlação (varia de -1 a 1).
- `p_valor_satisfacao`: Valor-p, que indica a significância estatística da correlação.
"""
correlacao_satisfacao, p_valor_satisfacao = pearson(produtos_vendas_avaliacao["Quantidade de Vendas"], produtos_vendas_avaliacao["Satisfação do Cliente"])
print(f"Correlação entre Vendas e Satisfação: {correlacao_satisfacao:.2f}")

"""
Exibe os produtos mais vendidos e sua avaliação média, ordenados pela quantidade de vendas em ordem decrescente.
"""
print("Produtos mais vendidos e sua avaliação média:")
print(produtos_vendas_avaliacao.sort_values(by="Quantidade de Vendas", ascending=False).to_string(index=False))
```

Pergunta: Alguma região domina as vendas de certos produtos?


```python
"""
Agrupa o DataFrame pelas colunas 'Região' e 'Categoria do Produto' e calcula a soma da coluna 'Quantidade de Vendas' para cada grupo.
O método `reset_index()` transforma o resultado em um DataFrame.
"""
dominancia_regiao = df.groupby(["Região", "Categoria do Produto"])["Quantidade de Vendas"].sum().reset_index()

"""
Identifica a região dominante (com maior quantidade de vendas) para cada categoria de produto.
- O método `groupby` agrupa os dados por 'Categoria do Produto'.
- O método `idxmax` retorna o índice da linha com o maior valor de 'Quantidade de Vendas' para cada grupo.
- O método `loc` é usado para selecionar as linhas correspondentes.
"""
dominancia_regiao = dominancia_regiao.loc[dominancia_regiao.groupby("Categoria do Produto")["Quantidade de Vendas"].idxmax()]

"""
Cria um gráfico de barras para visualizar as regiões dominantes nas vendas de cada categoria de produto.
- O eixo X representa as categorias de produtos.
- O eixo Y representa a quantidade de vendas.
- A cor das barras representa a região dominante.
- O título do gráfico é "Regiões Dominantes nas Vendas de Produtos".
- Os rótulos do eixo X são rotacionados em 45 graus para melhor visualização.
- A legenda é adicionada para identificar as regiões.
"""
plt.figure(figsize=(12,6))
sns.barplot(data=dominancia_regiao, x="Categoria do Produto", y="Quantidade de Vendas", hue="Região")
plt.xlabel("Produto")
plt.ylabel("Quantidade de Vendas")
plt.title("Regiões Dominantes nas Vendas de Produtos")
plt.legend(title="Região")
plt.show()

"""
Exibe a região dominante (com maior quantidade de vendas) para cada categoria de produto.
"""
print("Região dominante por produto:")
print(dominancia_regiao.to_string(index=False))
```

Pergunta: Alguma categoria está crescendo ou diminuindo em vendas ao longo do tempo?


```python
"""
Agrupa o DataFrame pela coluna 'Data' (convertida para o período mensal) e pela coluna 'Categoria do Produto'.
Calcula a soma da coluna 'Quantidade de Vendas' para cada grupo.
O método `reset_index()` transforma o resultado em um DataFrame.
"""
categoria_vendas_tempo = df.groupby([df["Data"].dt.to_period("M"), "Categoria do Produto"])["Quantidade de Vendas"].sum().reset_index()

"""
Converte a coluna 'Data' (que está no formato de período) para strings, facilitando a plotagem.
"""
categoria_vendas_tempo["Data"] = categoria_vendas_tempo["Data"].astype(str)

"""
Cria um gráfico de linha para visualizar a tendência de vendas por categoria ao longo do tempo.
- O eixo X representa as datas.
- O eixo Y representa a quantidade de vendas.
- Cada linha representa uma categoria de produto.
- O título do gráfico é "Tendência de Vendas por Categoria ao Longo do Tempo".
- Os rótulos do eixo X são rotacionados em 45 graus para melhor visualização.
- A legenda é adicionada para identificar as categorias.
"""
plt.figure(figsize=(12,6))
for categoria in categoria_vendas_tempo["Categoria do Produto"].unique():
    dados_categoria = categoria_vendas_tempo[categoria_vendas_tempo["Categoria do Produto"] == categoria]
    plt.plot(dados_categoria["Data"], dados_categoria["Quantidade de Vendas"], label=categoria)

plt.xlabel("Data")
plt.ylabel("Quantidade de Vendas")
plt.title("Tendência de Vendas por Categoria ao Longo do Tempo")
plt.xticks(rotation=45)
plt.legend(title="Categoria")
plt.show()

"""
Exibe o DataFrame `categoria_vendas_tempo`, que contém a tendência de vendas por categoria ao longo do tempo.
"""
print("Tendência de vendas por categoria ao longo do tempo:")
print(categoria_vendas_tempo.to_string(index=False))
```
